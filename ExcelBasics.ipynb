1. Explain the difference between Relative, Absolute and Mixed Cell Referencing.
  - **RELATIVE CELL REFERENCING** -> The cell address changes automatically when you copy the formula to another cell (EX- A1).
 - **ABSOLUTE CELL REFERENCING** -> The cell address remains fixed , no matter where the formula is copied. It uses dollar signs (Ex- $A$1).
 - **MIXED CELL REFERENCING** -> Either the row or the column is fixed, while the other remains relative (Ex- $A1 or A$1).

3. Using the data below, write a formula to calculate the average sales of items priced above 100 but less than 300.

| **Item** | **Price** | **Sales** |
|----------|-----------|-----------|
|Item A    | 90        | 1000      |
|Item B    | 150       | 1200      |
|Item C    | 250       | 1500      |

 - To calculate the average sales of items with a price between 100 and 300, use the **AVERAGEIFS** formula :

 =AVERAGEIFS(D3:D5, C3:C5, ">100", C3:C5, "<300")

 Where, **D3:D5** = The range containing the sales numbers you want to average.

 **C3:C5**, **">100"** = Only includes items where the price is greater than 100.

 **C3:C5, "<300"** = Further limits it to items where the price is also less than 300.
